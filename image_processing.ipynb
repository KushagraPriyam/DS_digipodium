### batch processing

```python
import os

image_list=[]
for filename in os.listdir('c:/users/kusha/downloads'):
    if filename.endswith('.jpg') or filename.endswith('.png'):
        #print(filename)
        image_list.append('c:/users/kusha/downloads/' + filename)
print('total images',len(image_list))      
for path in image_list:
    try:
        im=Image.open(path)
        im=im.convert('L')
        draw=ImageDraw(im)
        w,h=im.size
        f1=ImageFont.truetype('font1.ttf',100)
        draw.text((w//2,h//2),'image',fill='white',font=f1,anchor='mm')
        save_dir='images'
        os.makedirs(save_dir,exist_ok=True)
        filename=os.path.basename(path)
        im.save(f'{save_dir}/{filename}')
    except:
        pass
    
```
